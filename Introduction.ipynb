Why is there a need to analyze Algorithms. How do we constuct the paradigm of thinking for Big O notation. 
If we have to objectively compare algorithms we can consider space they occupy in memory and time taken for them to run. 



```python
def sum1(n):
    total_sum=0
    for i in range(0,n+1):
        total_sum += i
    return total_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return n*(n+1)/2
```

```python
sum2(10)
```

```python
%timeit (sum1(100))
```

```python
%timeit (sum2(100))
```

the way timeit works is that it will perform more loops if the program is faster, less loops to get the average if the program is slower, that way you aren't waiting forever to get an average speed for slower programs. Running at much faster rate, second algorithm is a lot more efficient. But the problem is we cant use time to run as an objective measurement. Because that will depend on speed of computer itself and hardware capabilities. we need an objective method to compare run times of algorithms and space occupied in memory irrespective of hardware(time and space). Big O is the solution. Compares algorithms relative to the size of the input/problem. Functions can take different run times but Big O notation thats relative or in terms of input then we can compare algorthms objectively for any input. We can see that when we formalize the notation. 

Big O describes how quickly run time will grow relative to the input as input grows arbitrarily large. Not run time but growth of run times as input grows( big O relative to input). as n gets arbitrarily large we are only interested in the terms that grow fastest as n gets large. 
Big O also known as asymptotic analysis. Big O notation- O(performance inside of these parenthesis)
Constant(1), Linear (n), logarithmic(log(n)), nlogn log linear, n^2 quadratic, n^3 cubic, 2^n exponential



```python
import matplotlib.pyplot as plt
from math import log
import numpy as np
%matplotlib inline
plt.style.use('bmh')

n=np.linspace(1,10,1000)
labels=['Constant','Logarithmic','Linear', 'Log Linear', 'Quadratic','Cubic','Exponential']
big_o= [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]
plt.figure(figsize=(8,8))
plt.ylim(0,50)
for i in range(len(big_o)):
    plt.plot(n,big_o[i],label=labels[i])

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
plt.title('Growth of Relative run time vs Size of the input')
```

There is a huge difference that Big O effieciency can make for the same value n against some run time.
